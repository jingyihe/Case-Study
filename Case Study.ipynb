#### 3. duplicate features
I noticed that some features have same values. In order to identify duplicate features, we can compare feature's quantiles (like feature clustering). But for simplicity, I use Cramer's V statistics (which can be used to calculate correlations between categorical variables, and notice the modality are samll for all the features) to identify multicollinearity to achieve the same goal. 

```python
import scipy.stats as stat

def chitest(confusion_matrix):
    return stat.chi2_contingency(confusion_matrix)[0]

def cramers_stat(confusion_matrix):
    chi2 = stat.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    return np.sqrt(chi2 / (n*(min(confusion_matrix.shape)-1)))

def cramers_stat_correct(confusion_matrix):
    chi2 = stat.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))    
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min( (kcorr-1), (rcorr-1)))
```

```python
rows_list = []
for i in range(len(cols)):
    for j in range(i+1,len(cols)):
        confusion_matrix = pd.crosstab(train_df[cols[i]].astype(str), train_df[cols[j]].astype(str))
        rows_list.append([cols[i],cols[j],chitest(confusion_matrix), cramers_stat(confusion_matrix), cramers_stat_correct(confusion_matrix)])
chitest = pd.DataFrame(rows_list)  
chitest.columns=["cur_features","compar_features","Chi2","Cramers V statistic","Adjust Cramers V statistic"]
chitest.sort_values(['Adjust Cramers V statistic'],ascending=False)
```

```python
# also tried 0.8. The overall result does not change much
feature_to_remove = chitest[chitest['Adjust Cramers V statistic']>0.75]["compar_features"].unique()
feature_to_remove
```
